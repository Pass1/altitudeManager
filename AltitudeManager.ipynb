```python
out_filename = 'C:/Users/FalconViz/Documents/DarProject/intermediate_file.waypoints'

search_locations = 2
cells_delta = 4
desired_altitude = 100
altis = []
#relative alti
#wpFormat = '{:d}\t0\t3\t16\t0.000000\t0.000000\t0.000000\t0.000000\t{:.6f}\t{:.6f}\t{:.6f}\t1\n'
#absolute alti
wpFormat = '{:d}\t0\t0\t16\t0.000000\t0.000000\t0.000000\t0.000000\t{:.6f}\t{:.6f}\t{:.6f}\t1\n'
target = open(out_filename, 'w')

target.write('QGC WPL 110\n')
#set home
target.write('0\t0\t0\t16\t0.000000\t0.000000\t0.000000\t0.000000\t21.359465\t39.713267\t100.000000\t1\n')
#set speed
target.write('1\t0\t0\t178\t1.000000\t15.000000\t0.000000\t0.000000\t0.000000\t0.000000\t0.000000\t1\n')
target.write('2\t0\t0\t206\t13.778950\t0.000000\t0.000000\t0.000000\t0.000000\t0.000000\t0.000000\t1\n')
cmdCounter = 3

x_cell = int(round((waypoints[0,9]-xllcorner)/cellsize))
y_cell = nrows-int(round((waypoints[0,8]-yllcorner)/cellsize))
#previous_alti = 100 + np.max(altitude[y_cell-search_locations:y_cell+search_locations,
#                                      x_cell-search_locations:x_cell+search_locations])
for i in range(1,waypoints.shape[0]):
    x_cell = int(round((waypoints[i-1,9]-xllcorner)/cellsize))
    y_cell = nrows-int(round((waypoints[i-1,8]-yllcorner)/cellsize))
    #print('Altitudes:', altitude[y_cell-search_locations:y_cell+search_locations, x_cell-search_locations:x_cell+search_locations])
    alti = 100 + np.max(altitude[y_cell-search_locations:y_cell+search_locations,
                                 x_cell-search_locations:x_cell+search_locations])
    print('To file', alti)
    dist = np.sqrt((waypoints[i, 8] - waypoints[i-1, 8]) ** 2 + (waypoints[i, 9] - waypoints[i-1, 9]) ** 2)


    target.write(wpFormat.format(cmdCounter, 
                                 waypoints[i-1, 8],
                                 waypoints[i-1, 9], 
                                 alti))
    
    cmdCounter = cmdCounter + 1
    
    y_dist = waypoints[i, 8] - waypoints[i-1, 8]
    x_dist = waypoints[i, 9] - waypoints[i-1, 9]
    n_cells = dist // cellsize
    y_ratio = y_dist / n_cells
    x_ratio = x_dist / n_cells
    previous_alti = alti
    if (dist / cellsize ) > cells_delta:
        for sc in range(cells_delta, int(dist / cellsize), cells_delta):
            x_long = sc*x_ratio+waypoints[i-1,9]
            #x_cell = int(round((x_long-xllcorner)/cellsize))
            y_lat = sc*y_ratio+waypoints[i-1,8]
            #y_cell = nrows - int(round((y_lat-yllcorner)/cellsize))
            
            x_cell = int(round((x_long-xllcorner)/cellsize))
            y_cell = nrows-int(round((y_lat-yllcorner)/cellsize))
            #plt.scatter(x_cell, nrows-y_cell)
            #print(np.max(altitude[x_cell-search_locations:x_cell+search_locations,
            #                      y_cell-search_locations:y_cell+search_locations]))
            alti = 100 + np.max(altitude[y_cell-search_locations:y_cell+search_locations,
                                        x_cell-search_locations:x_cell+search_locations])
            #adjust for maximum descent rate
            #if alti < previous_alti:
                #print(alti)
            #    alti = np.max([alti, previous_alti - (dist*111110 / 4.5)])
                #print(alti)
            #print(x_long, y_lat, altitude[y_cell,x_cell], np.max(altitude[y_cell-search_locations:y_cell+search_locations,x_cell-search_locations:x_cell+search_locations]))
            target.write(wpFormat.format(cmdCounter, y_lat, x_long, alti))
            cmdCounter = cmdCounter + 1
            

target.close()
print('Done')
```
